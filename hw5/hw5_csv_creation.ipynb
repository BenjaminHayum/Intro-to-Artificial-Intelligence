```python
with open('Mendota_Ice.html') as html:
    string = html.read()
```

```python
import re
import pandas as pd
```

```python
re.findall("x", "xxxx")
```

```python
print(string)
```

```python
pre_years = re.findall("(\d\d\d\d)-\d\d<BR>|(\")<BR>|(\d\d\d\d)-\d\d<br>|(\")<br>|(\d\d\d\d)-\d\d</FONT>", string)
pre_years
```

```python
years = list()
for pre_year in pre_years:
    for entry in pre_year:
        if len(entry) != 0:
            years.append(entry)
years
```

```python
pre_days = re.findall("[^-](\d\d\d|\d\d)<BR>|(-+)<BR>|[^-](\d\d\d|\d\d)<br>|(-+)<br>|[^-](\d\d\d|\d\d)\n<br>|[^-](\d\d\d|\d\d)</FONT>", string)
# To cut off the 4 "---"s at the beginning
pre_days
```

```python
days = list()
for pre_day in pre_days:
    for entry in pre_day:
        if len(entry) != 0:
            days.append(entry)
days = days[4:]
days
```

```python
print(len(years), len(days))
```

```python
# Cutting off anything before 1855
years = years[3:]
days = days[3:]
```

```python
# Cutting off anything after 2020
years = years[:-1]
days = days[:-1]
```

```python
print(len(years), len(days))
```

```python
for i in range(len(days)):
    print(years[i], days[i])
```

```python
post_years = list()
post_days = list()

skip_boolean = False
for i in range(len(years)):
    if skip_boolean == True:
        skip_boolean = False
        continue

    if days[i] == "-" or days[i] == "--" or days[i] == "---":
        if years[i+1] == "\"":
            post_years.append(years[i])
            post_days.append(days[i+1])
            skip_boolean = True
    else:
        post_years.append(years[i])
        post_days.append(days[i])
```

```python
for i in range(len(post_years)):
    print(post_years[i], post_days[i])
```

```python
data = dict()
data["year"] = post_years
data["days"] = post_days
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('hw5.csv', index = False)
```
